```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as hc
```

```python
##import the data
```

```python
data = pd.read_csv("datasets/Mall_Customers.csv")
```

```python
data.head()
```

```python
X = data.iloc[:,[3,4]].values
```

```python
## building the dendrograms
```

```python
dg = hc.dendrogram(hc.linkage(X,method='ward'))
plt.title("dendrogram")
plt.xlabel("clusters")
plt.ylabel("euclidian distinces")
plt.show()
```

```python
## training the hirerachical cluster algo with data
```

```python
aglo = AgglomerativeClustering(n_clusters=5,linkage='ward',affinity='euclidean') #ward is minimum variance
y_hc = aglo.fit_predict(X)
```

```python
## create a visuals
```

```python
plt.scatter(X[y_hc==0,0], X[y_hc==0,1],s=100,c='red',label='cluster1')
plt.scatter(X[y_hc==1,0], X[y_hc==1,1],s=100,c='green',label='cluster2')
plt.scatter(X[y_hc==2,0], X[y_hc==2,1],s=100,c='blue',label='cluster3')
plt.scatter(X[y_hc==3,0], X[y_hc==3,1],s=100,c='yellow',label='cluster4')
plt.scatter(X[y_hc==4,0], X[y_hc==4,1],s=100,c='purple',label='cluster5')
plt.title("Clusters of customers")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
